# Mean : $$ \overline x = \sum_{i=1}^n \frac{x_i }{n}   $$
    
    - Central Value of the data 
    - Calculated by Sum of all data , divided by total number of data 
    
# Median :
    - value separating the higher half from the lower half of a data
    
# Mode : 
    - Most repeated value in the data

```python
%matplotlib inline


# Imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mode
import seaborn as sb
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize

scaler = StandardScaler()
normaliser = MinMaxScaler()
```

# Data visual with mean , median and mode 

```python
# Data 
Data = np.random.randint(low=10, high=100, size=100)
# print(Data)

mean = Data.mean()
median = np.median(Data)
mode_value = mode(Data)
# print("Mean : {}".format(mean))
# print("Median : {}".format(median))
# print("Mode : {}".format(mode_value))


#Histogram
plt.figure(figsize=(20, 10))
plt.hist(x=Data, color='c', edgecolor='k', alpha=0.65)
plt.axvline(x=mean, linestyle=':', color = 'k')
plt.axvline(x=median, linestyle='--', color = 'c')
plt.axvline(x=mode_value.mode, linestyle='-.', color='y')

# Label and Titles

plt.xlabel('Data', size=20)
plt.ylabel('Count', size=20)
plt.title('Mean : {}, Median : {}, Mode : {}'.format(mean, median, mode_value), size=20)
plt.gca().legend(('mean', 'median', 'mode_value'))
sb.set_style('whitegrid')
```

# Mean, Median, Mode after Normalisation

```python
data = scaler.fit_transform(Data.reshape(-1, 1))
# print(data)

mean = data.mean()
median = np.median(data)
mode_value = mode(data)
print("Mean : {}".format(mean))
print("Median : {}".format(median))
print("Mode : {}".format(mode_value))


#Histogram
plt.figure(figsize=(20, 10))
plt.hist(x=data, color='c', edgecolor='k', alpha=0.65)
plt.axvline(x=mean, linestyle=':', color = 'k')
plt.axvline(x=median, linestyle='--', color = 'c')
plt.axvline(x=mode_value.mode, linestyle='-.', color='y')

# Label and Titles

plt.xlabel('Data', size=20)
plt.ylabel('Count', size=20)
plt.title('Mean : {}, Median : {}, Mode : {}'.format(mean, median, mode_value), size=20)
plt.gca().legend(('mean', 'median', 'mode_value'))
sb.set_style('whitegrid')
```
