# Mean : $$ \overline x = \sum_{i=1}^n \frac{x_i }{n}   $$
    
    - Central Value of the data 
    - Calculated by Sum of all data , divided by total number of data 
    
# Median :
    - value separating the higher half from the lower half of a data
    
# Mode : 
    - Most repeated value in the data

```python
%matplotlib inline


# Imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mode
import seaborn as sb
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize

scaler = StandardScaler()
normaliser = MinMaxScaler()
```

# Data visual with mean , median and mode 

```python
# Defines:
hist_background = 'wheat'
edge_colour = 'black'
mean_color = 'red'
median_color = 'black'
mode_color = 'green'
line_style = '-.'
fig_size = (15, 5)


def mean_mode_median(Data):
    Mean = np.round(np.mean(Data), decimals=3)
    Median = np.round(np.median(Data), decimals=3)
    Mode = np.round(mode(Data).mode, decimals=3)
    print("Mean : {}".format(Mean))
    print("Median : {}".format(Median))
    print("Mode : {}".format(Mode))
    return Mean, Median, Mode


def show_hist(Data):
    Mean, Median, Mode = mean_mode_median(Data)
    #Histogram
    plt.figure(figsize=fig_size)
    plt.hist(x=Data, color=hist_background, edgecolor=edge_colour, alpha=0.65)
    plt.axvline(x=Mean, linestyle=line_style, color = mean_color)
    plt.axvline(x=Median, linestyle=line_style, color = median_color)
    plt.axvline(x=Mode, linestyle=line_style, color=mode_color)
    
    # Label and Titles
    plt.xlabel('Data Value', size=20)
    plt.ylabel('Data Frequency', size=20)
    plt.title('Mean : {}, Median : {}, Mode : {}'.format(Mean, Median, Mode), size=20)
    plt.gca().legend(('mean', 'median', 'mode_value'))
    sb.set_style('whitegrid')
```

```python
# Data 
Data = np.random.randint(low=10, high=100, size=100)
show_hist(Data)
```

# Mean, Median, Mode after Standardisation

```python
data = scaler.fit_transform(Data.reshape(-1, 1))
show_hist(data)
```

# Mean, Median, Mode after Normalization

```python
data = normaliser.fit_transform(Data.reshape(-1, 1))
show_hist(data)
```
